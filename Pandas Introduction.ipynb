### Get Dummies

```python
f={'firstname':['Arya','Bhuvan','Raghu','Karthish','Hema','sowmya'],
       'lastname':['Kumar','Jakob','Raghavan','Singh','Khan','Khan'],
       'worktype':['Service','Business','Student','Service','Business','Business'],
       'country':['India','USA','USA','Sweden','Australia','Germany']}
e=pd.DataFrame(f,columns=['firstname','lastname','worktype','country'])
print(e)
```

```python
g=pd.get_dummies(e['worktype'])
h=pd.get_dummies(e['country'])
print(g)
print(h)
```

```python
e
```

```python
frames=[e,g,h]
result=pd.concat(frames,axis=1)
result
```

### concatination

```python
india_weather=pd.DataFrame({
    'city':['mumbai','delhi','chennai'],
    'temperature':[32,45,33],
    'humidity':[70,65,75]
})
india_weather
```

```python
usa_weather=pd.DataFrame({
    'city':['california','sanfransisco','newyork'],
    'temperature':[38,47,53],
    'humidity':[78,55,85]
})
usa_weather
```

```python
df=pd.concat([india_weather,usa_weather])
df
```

```python
#ignore index
df=pd.concat([india_weather,usa_weather],ignore_index=True)
df
```

```python
#concatinate using keys
df=pd.concat([india_weather,usa_weather],keys=['india','usa'])
df
```
